```python
from PIL import Image, ImageFilter
```

```python
# Opening the image (R prefixed to string
# in order to deal with '\' in paths)
image1 = Image.open(r"mehul.jpeg")
image2 = Image.open(r"beard-face.jpg")
image3 = Image.open(r"transparent-face-mask.jpg")
```

```python

# Converting the image to greyscale, as edge detection
# requires input image to be of mode = Greyscale (L)
image1 = image1.convert("L")
image2 = image2.convert("L")
image3 = image3.convert("L")
```

```python

# Detecting Edges on the Image using the argument ImageFilter.FIND_EDGES
image1 = image1.filter(ImageFilter.FIND_EDGES)
image2 = image2.filter(ImageFilter.FIND_EDGES)
image3 = image3.filter(ImageFilter.FIND_EDGES)
```
